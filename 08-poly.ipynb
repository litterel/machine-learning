```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
x = np.linspace(-3,3,100)
inx = np.argsort(x)
x = x[inx]
y = x**2 + x +3 + np.random.normal(0,1,100)
X = x.reshape(-1,1)


xx = np.linspace(-3,3,100)
XX = xx.reshape(-1,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly = PolynomialFeatures(30)
poly.fit(x.reshape(-1,1))
X = poly.transform(x.reshape(-1,1))
#poly.fit(xx.reshape(-1,1))
XX = poly.transform(xx.reshape(-1,1))

lin_reg = LinearRegression()
lin_reg.fit(X,y)
yy = lin_reg.predict(X)
lin_reg.score(X,y)
```
