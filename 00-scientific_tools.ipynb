## 偏度，峰度

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
a = np.random.randn(10000)
b = np.random.uniform(low=-4.,high=4.,size=10000)
c = a+b
```

```python
sns.distplot(c,bins=50)
```

```python
stats.gaussian_kde(a)
```
