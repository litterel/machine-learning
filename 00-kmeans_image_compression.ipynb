```python
from skimage import io
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = io.imread('sample.jpeg')
```

```python
plt.imshow(data)
```

```python
rows = data.shape[0]
cols = data.shape[1]
```

```python
X = data.reshape(rows*cols,data.shape[2])
```

```python
km = KMeans(n_clusters=128,n_jobs=-1)
```

```python
km.fit(X)
```

```python
res = km.predict(X)
```

```python
plt.imshow(res.reshape(rows,cols))
```
